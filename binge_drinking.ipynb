```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
bd_file = "binge_drinking.csv"
bd_df = pd.read_csv(bd_file)
bd_df.head()
```

```python
bd_df = bd_df[['Year', 'LocationAbbr', 'LocationDesc', 'Topic', 'Response', 'Data_Value_Unit', 'Data_Value_Type', 'Data_Value', 'Sample_Size']].copy()
bd_df.head()
```

```python
bd_df = bd_df.rename(columns = {"Year": "year", "LocationAbbr": "location_abbr", "LocationDesc": "location", "Topic": "topic", "Response": "response", "Data_Value_Unit": "value_unit", "Data_Value_Type": "value_type", "Data_Value": "data_value", "Sample_Size": "sample_size"})
bd_df.head()
```

```python
import psycopg2
rds_connection_string = "postgres:postgres@localhost:5432/alcohol_consumption_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
bd_df.to_sql(name='binge_drinking', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from binge_drinking', con=engine).head()
```
